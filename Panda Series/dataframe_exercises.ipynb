Pandas DataFrame Exercises

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
%matplotlib inline

print(pd.__version__)
```

```python
# step1 : creat an empty pandas data frame

pd.DataFrame(data=[None],
             index=[None],
             columns=[None])
```

```python
# given marvel data

marvel_data = [
    ['Spider-Man', 'male', 1962],
    ['Captain America', 'male', 1941],
    ['Wolverine', 'male', 1974],
    ['Iron Man', 'male', 1963],
    ['Thor', 'male', 1963],
    ['Thing', 'male', 1961],
    ['Mister Fantastic', 'male', 1961],
    ['Hulk', 'male', 1962],
    ['Beast', 'male', 1963],
    ['Invisible Woman', 'female', 1961],
    ['Storm', 'female', 1975],
    ['Namor', 'male', 1939],
    ['Hawkeye', 'male', 1964],
    ['Daredevil', 'male', 1964],
    ['Doctor Strange', 'male', 1963],
    ['Hank Pym', 'male', 1962],
    ['Scarlet Witch', 'female', 1964],
    ['Wasp', 'female', 1963],
    ['Black Widow', 'female', 1964],
    ['Vision', 'male', 1968]
]
```

```python
# create a marvel_df pandas DataFrame with given data above

marvel_df = pd.DataFrame(data=marvel_data)

marvel_df
```

```python
# add column names to marvel_df

col_names = ['Name', 'Sex', 'First_Appearance']   # initialize column names
marvel_df.columns = col_names    #  store column names
marvel_df   # check
```

```python
# add index names to marvel_df

marvel_df.index = marvel_df['Name']
marvel_df
```

```python
# drop repeated name column since its index 

marvel_df = marvel_df.drop(['Name'], axis=1)   # axis 1 means column , 0 means row
marvel_df
```

```python
# drop 'Namor' and ' Hank Pym' rows

marvel_df = marvel_df.drop(['Namor', 'Hank Pym'], axis = 0)
marvel_df
```

```python
# show last 5 elements of marvel_df
marvel_df.iloc[-5:]
```

```python
# show first_appearance of all middle elements 

marvel_df.iloc[1:-1,].First_Appearance.to_frame()
```

```python
# show first and last elements on marvel_df 

marvel_df.iloc[[0, -1]]
```

DataFrame Manipulation and Operations

```python
# modify first appearance of Vision to year 1964

marvel_df.loc['Vision', 'First_Appearance'] = 1964
marvel_df
```

```python
# add new column called years since with years since first appearance 
marvel_df['Years_Since'] = 2018 - marvel_df['First_Appearance']
marvel_df
```

DataFrame Boolean Arrays - Masks 

```python
#  make a mask showing female characters
mask = marvel_df['Sex'] == 'female'
mask
```

```python
# get male characters

mask = marvel_df['Sex'] == 'male'
mask
```

```python
# get characters with first appearance after 1970
mask = marvel_df['First_Appearance'] < 1970
mask
```

```python
# get female characters with first appearance after 1970
mask = (marvel_df['Sex'] == 'female') & (marvel_df['First_Appearance'] > 1970)
mask
```
