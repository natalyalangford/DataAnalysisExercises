More Visualizations:
    - Using the basic API of matplotlib

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

#matplotlib inline
```

Global API
    - matplotlibs default APU has a global MATLAB-style interface 

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize=(12, 6))

plt.title('My Nice Plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
# creating a Matplotlib figure with two subplots
# each containing a plot of a quadratic function and a line

plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1)  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X Squared')
```

OOP Interface
    - Object oriented programming

```python
# creating a Matplotlib figure and one or more axes (subplots) in OOP manner

fig, axes = plt.subplots(figsize=(12, 6))

# figsize=(12, 6) -> sets size of figure to 12 in width and 6 in height
# use the axes objects to plot data onto specific subplots
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2')

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("My Nice Plot")

axes.legend()

fig
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '-og', label="solid green")
axes.plot(x, x + 1, '--c', label="dashed cyan")
axes.plot(x, x + 2, '-.b', label="dashdot blue")
axes.plot(x, x + 3, ':r', label="dotted red")

axes.set_title("My Nice Plot")

axes.legend()
```

2) Scatter Plot 

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2  # 0 to 15 point radii
```

```python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
```

4) KDE - Kernel Density Estimation

```python
from scipy import stats

# using SciPy's gaussian_kde function estimate the probability density function of a set of data values
# result-> density estimation function 
# use result  to compute the estimated PDF at specific points or to perform other density-related calculations

density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12, 6))

# create a range of x-values values2 using np.linspace
# range -> extends from the min val minus 10 to the max val plus 10 and contains 100 evenly spaced points
# this range is used for plotting the KDE curve
values2 = np.linspace(min(values)-10, max(values)+10, 100)

#  plt.plot to plot the KDE curve 
#  density() computes the estimated PDF values for the range of x-values in values2
#  color='#FF7F00' sets the color of the plot to a shade of orange
plt.plot(values2, density(values2), color='#FF7F00') 

# fill the area under the KDE curve -> creates a shaded region that represents the density of the data
# alpha=0.5 controls the transparency of the filled area
# color='#FF7F00' sets the fill color to the same shade of orange
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')

# sets the x-axis limits-> only the portion of the KDE plot within the range from -5 to 5 is displayed
plt.xlim(xmin=-5, xmax=5)

#  displays the figure with the KDE plot.
plt.show()
```

5) Combine Plots

6) Bar Plots

```python
# generate two random arrays Y and Y2 containing five random numbers each

# np.random.rand(1, 5) generates a random array of shape (1, 5)
# [0] is used to access the first (and only) row in each array
# converting it to a 1-dimensional (1D) array or simply a list of random numbers
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
# use values stored in the array Y to create a bar chart

# create a figure with a size of 12 in width and 4 in height
plt.figure(figsize=(12, 4))

# set the width of the bars
barWidth = 0.5

# create bar plot
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')

# display plot

plt.show()
```

```python
# can be stacked bars
# add a legend to the plot:

# create a figure with a size of 12 in width and 4 in height
plt.figure(figsize=(12, 4))

# set width of bars
barWidth = 0.5

# create first set of bars using plt.bar for Y
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894', label='Label Y')

# create second set of bars using plt.bar for Y2, with Y as the 'bottom' parameter
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e17055', bottom=Y, label='Label Y2')

# add a legend to distinguish between the two sets of bars
plt.legend()

# display figure containing the grouped bar plot
plt.show()
```
